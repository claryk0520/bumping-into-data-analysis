```python
data_s = pd.DataFrame({'음식명' : ['돈가스', '우동', '냉면'],
                      '평점' : [4.3, 4.2, 4.6],
                      '가격' : [10000, 9000, 12000],
                      '위치' : ['삼성동', '명동', '을지로입구']})
data_s
```

```python
data_s.melt('음식명')
```

```python
data_s.melt(id_vars=['음식명'], value_vars=['가격', '위치'], var_name='항목명', value_name='항목값')
```

```python
data_s.melt('음식명').pivot('음식명', 'variable', 'value')
```

```python
data_s.melt('음식명').pivot('음식명', 'variable', 'value').reset_index()
```
