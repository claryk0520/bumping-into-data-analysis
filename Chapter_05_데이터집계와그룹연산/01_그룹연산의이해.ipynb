```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
len(tips)
```

```python
tips['tip_percent'] = tips['tip'] / tips['total_bill']
```

```python
tips.head()
```

```python
tips_byday = tips.groupby(['day', 'smoker'])
```

```python
tips_byday_percent = tips_byday['tip_percent']
```

```python
tips_byday_percent.agg('mean')
```

```python
tips_byday_percent.mean()
```

```python
tips_byday_percent.agg(['mean', 'std', 'sum'])
```

```python
tips_byday_percent.agg([('팁_평균', 'mean'), ('팁_표준편차', 'std')])
```

```python
function_list = ['count', 'mean', 'max']
```

```python
tips_byday['tip_percent', 'total_bill'].agg(function_list)
```

```python
function_tuple = [('팁_평균', 'mean'), ('팁_표준편차', 'std')]
```

```python
tips_byday['tip_percent', 'total_bill'].agg(function_tuple)
```

```python
tips_byday.agg({'tip' : 'max', 'size' : 'sum'})
```

```python
tips_byday.agg({'tip_percent' : ['min', 'max', 'mean', 'std'], 'size' : 'sum'})
```

```python
tips.groupby(['day', 'smoker']).mean()
```

```python
tips.groupby(['day', 'smoker'], as_index=False).mean()
```

```python
tips.groupby(['day', 'smoker']).mean().reset_index()
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins = penguins.dropna()
```

```python
penguins['body_mass_g_bin'] = pd.cut(penguins['body_mass_g'], bins=5)
```

```python
penguins.head()
```

```python
penguins.body_mass_g_bin.value_counts()
```

```python
penguins['flipper_length_mm_bin'] = penguins.groupby('body_mass_g_bin').flipper_length_mm.apply(pd.cut, bins=2)
```

```python
# 위 코드와 동일한 결과
penguins['flipper_length_mm_bin'] = penguins.groupby('body_mass_g_bin')['flipper_length_mm'].transform(lambda x: pd.cut(x, bins=2)) 
```

```python
penguins.head()
```

```python
penguins.flipper_length_mm_bin.value_counts()
```

```python
penguins.groupby('island').bill_length_mm.apply(pd.cut, bins=2)
```

```python
penguins['bill_length_mm_bin_by_island'] = penguins.groupby('island').bill_length_mm.apply(pd.cut, bins=2)
```

```python
# 위 코드와 동일한 결과
penguins['bill_length_mm_bin_by_island'] = penguins.groupby('island')['bill_length_mm'].transform(lambda x: pd.cut(x, bins=2))
```

```python
penguins[['island', 'bill_length_mm', 'bill_length_mm_bin_by_island']].head()
```

```python
penguins[['island', 'bill_length_mm', 'bill_length_mm_bin_by_island']].tail()
```

```python
penguins.bill_length_mm_bin_by_island.value_counts()
```
